```python
def knn(mnist_train,mnist_valid,Y,deb,fin):
    result= pd.DataFrame()
    neighbour=[]
    score=[]
    for i in range (1,31):
        mnist_train.columns = mnist_train.columns.astype(str)
        mnist_valid.columns = mnist_valid.columns.astype(str)
        n_n = KNeighborsClassifier(n_neighbors=i)
        neighbour.append(n_n.fit(mnist_train.iloc[:,deb:fin],mnist_train[Y]))
        n_n_score_valid=n_n.score(mnist_valid.iloc[:,deb:fin], mnist_valid[Y])
        score.append(n_n_score_valid)

    result['score validation']= score
    result['neighbour']= neighbour
    best_index = result['score validation'].idxmax()
    best_neighbour= result['neighbour'][best_index]
    knn_final=best_neighbour.fit(mnist_train.iloc[:,deb:fin],mnist_train.label)
    return result,knn_final

test=knn(mnist_train,mnist_valid,'label',1,mnist.shape[1])
test
```

```python
knn_error_gen=1-test[1].score(mnist_test.iloc[:,1:mnist.shape[1]], mnist_test.label)
knn_error_gen
```

```python
def my_hog(row, ori, cell):
    return(pd.Series(hog(row.iloc[1:].to_numpy().reshape(28,28,1), orientations=ori, pixels_per_cell=(cell, cell), cells_per_block=(1,1),channel_axis=2)))
```

```python
hog_train = mnist_train.apply(my_hog, axis=1, args=(12,7))
hog_valid= mnist_valid.apply(my_hog, axis=1, args=(12,7))
hog_test = mnist_test.apply(my_hog, axis=1, args=(12,7))
hog_train['label'] = mnist_train.label
hog_valid['label'] = mnist_valid.label
hog_test['label'] = mnist_test.label
hog_test
```

```python
test=knn(hog_train,hog_valid,'label',0,192)
test
#Pourquoi 192 et pas 193 ? hog_train.shape[1]
```
